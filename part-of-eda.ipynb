```python
import numpy as np
import pandas as pd
```

```python
train_labels = pd.read_csv("/kaggle/input/image-matching-challenge-2024/train/train_labels.csv")
train_labels.head()
```

```python
def str_to_arr(string: str, shape: tuple[int,int]):
    lst = string.split(';')
    array = []
    for s in lst:
        array.append(float(s))
    return np.array(array).reshape(shape)
```

```python
for i, row in train_labels.iterrows():
    _, rotation_matrix, translation_vector, calibration_matrix, __, ___ = row
    rotation_matrix = str_to_arr(rotation_matrix,(3,3))
    translation_vector = str_to_arr(translation_vector,(3,1))
    calibration_matrix = str_to_arr(calibration_matrix,(3,3))
    camera_matrix = np.dot(calibration_matrix, np.hstack((rotation_matrix, translation_vector)))
    print(camera_matrix)
```
